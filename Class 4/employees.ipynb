```python
import pandas as pd
```

```python
df= pd.DataFrame(pd.read_csv("employees.csv"))
df
```

```python
male=df[df['sex'] == 'Male']
female=df[df['sex'] == 'Female']
avg_f=female.loc[:,"age"].mean()
avg_m= male.loc[:,"age"].mean()
series= pd.Series([avg_f,avg_m])
series.index=["female","male"]
series.plot.pie()
```

```python
male=df[df['sex'] == 'Male']
female=df[df['sex'] == 'Female']
salary_f=female.loc[:,"salary"]
salary_m= male.loc[:,"salary"]
salary_m.plot.hist()
salary_f.plot.hist()
```
