```python
import pandas as pd

krk_airlines = pd.read_csv('krk-airlines.csv')
krk_flights = pd.read_csv('krk-flights.csv')
krk_passengers = pd.read_csv('krk-passengers.csv')
```

```python
flight_dest = krk_flights.loc[:,['flight', 'to']]
flight_dest
```

```python
flight_airline_aircraft = pd.merge(krk_flights,krk_airlines, on='airlineid')
flight_airline_aircraft = flight_airline_aircraft.loc[:,['flight', 'airline', 'airplane']]
flight_airline_aircraft
```

```python
flight_passenger = pd.merge(krk_flights, krk_passengers, on='flight')
flight_passenger.query("to == 'London'").loc[:,['name', 'surname']]
```

```python
flight_passenger.query("to == 'Paris' and gender == 'Female'").loc[:,['name', 'surname']].sort_values(by='surname')
```

```python
flight_passenger.query("to =='Berlin'").loc[:,'to'].count()
```

```python
flight_passenger.query("gender == 'Male'").loc[:,'from'].count()
```
